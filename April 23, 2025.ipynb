## Support Vector Machine (SVM)

SVM : Used for (Classification and Regression) but we more use for (classification)
Mostly used for classification problem

Kernal : transformation function

Types of kernal : Linear, RBF - Gaussian, Polynomial



```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler  #convert distribution of data in such way mean is 0 and variance is 1
from sklearn.svm import SVC
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
```

```python
iris = datasets.load_iris()
```

```python
x = iris.data
```

```python
#target data
y = iris.target
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state= 42)
```

```python
scaler = StandardScaler()                               # convert data using z tansform (0 to 1)
```

```python
x_train = scaler.fit_transform(x_train)
x_test = scaler.fit_transform(x_test)
```

```python
svm_model = SVC()                                       # Support vector classification
```

```python
svm_model.fit(x_train,y_train)                          # Train the x and y data
```

```python
y_pred = svm_model.predict(x_test)                       # Give the prediction on test data
y_pred
```

```python
y_test                                                    # y Test data
```

```python
accuracy_score(y_test,y_pred)*100                        # Accuracy
```
